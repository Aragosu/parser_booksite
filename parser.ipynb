### Решение

```python
def parse_chitai_gorod(url):
    
    # импорт всего необходимого
    import time
    from tqdm import tqdm
    import requests
    from bs4 import BeautifulSoup
    from random import randint,uniform,choices
    
    # начальная ссылка
    params = {'page': 1,
             'onlyAvailable':1}
    page = requests.get(url, params = params)
    soup = BeautifulSoup(page.text, 'html')
    
    # считываем количество страниц по ссылке
    list_of_page = [int(i) for i in range(1,int(soup.find_all('span', {'class': 'pagination__text'})[-1].get_text(strip=True))+1)]
    
    # перебираем сслыки на каждую книгу со всех страниц в список
    if len(list_of_page) <= 50:
        k_page = list_of_page
    else:
        k_page = choices(list_of_page, k=50) # если больше 50 страниц, то выбираем 50 страниц в случайном порадке
    
    urls = []
    for i in tqdm(k_page):    
        params_page = {'page': i,
                      'onlyAvailable':1}

        page_n = requests.get(url, params = params_page)
        soup_n = BeautifulSoup(page_n.text, 'html')

        for link in soup_n.find_all('a'):
            if link.get('href') is not None:
                if '/product/' in link.get('href'):
                    urls.append(link.get('href'))

        urls = list(set(urls)) # массив с уникальными ссылками на каждую книгу
    
    # проходимся по каждой странице и забираем название, цену, кол-во страниц книги
    page_list = []
    cnt = 0

    for i in tqdm(urls):
        cur_page = requests.get('https://www.chitai-gorod.ru'+i)
        cur_soup = BeautifulSoup(cur_page.text, 'html')
        
        # переодически вываливается ошибка 5xx (502/500), поэтому чтобы парсер не падал дописал это условие
        # понимаю, что таким образом вы не добавляем часть книг, но мы все равно не все книги парсим, а лишь выборку
        if cur_page.ok:
            # название
            name_book = cur_soup.find_all('h1', {'itemprop': 'name'})[0].get_text(strip=True)
            
            # стоимость - у некоторых книг нет стоимости
            if len(cur_soup.find_all('div', {'class': 'product-detail-offer-header__price'})) == 0:
                price_book = 0
            else:
                price_book = cur_soup.find_all('div', {'class': 'product-detail-offer-header__price'})[0].span.get('content')

            # кол-во страниц страниц - у некоторых книг нет кол-ва страниц
            if len(cur_soup.find_all('span', {'itemprop': 'numberOfPages'})) == 0:
                page_book = 0
            else:
                page_book = cur_soup.find_all('span', {'itemprop': 'numberOfPages'})[0].get_text(strip=True)

            cnt += 1
            
            # на каждой 10 книге врубаем рандомный сон
            if cnt%10 == 0:
                time.sleep(uniform(1,3))
        else:
            continue
        
        
        page_list.append([name_book,price_book,page_book])
    
    return page_list
```

```python
import pandas as pd

df_classic = pd.DataFrame(book_classic, columns = ['name','price','pages'])
df_classic['price'] = df_classic['price'].astype(int)
df_classic['pages'] = df_classic['pages'].astype(int)

df_python = pd.DataFrame(book_python, columns = ['name','price','pages'])
df_python['price'] = df_python['price'].astype(int)
df_python['pages'] = df_python['pages'].astype(int)

df_ml = pd.DataFrame(book_ml, columns = ['name','price','pages'])
df_ml['price'] = df_ml['price'].astype(int)
df_ml['pages'] = df_ml['pages'].astype(int)

df_c_plus_plus = pd.DataFrame(book_c_plus_plus, columns = ['name','price','pages'])
df_c_plus_plus['price'] = df_c_plus_plus['price'].astype(int)
df_c_plus_plus['pages'] = df_c_plus_plus['pages'].astype(int)
```

### 1. Какова в среднем длина (в страницах) книги из категории классическая литература в ассортименте магазина

```python
df_classic_mean = df_classic[df_classic.pages > 0].pages.mean()
df_classic_median = df_classic[df_classic.pages > 0].pages.median()

print(f'''Кол-во страниц по классике:
Средняя:{df_classic_mean}
Медиана:{df_classic_median}''')
```

### 2. Какое у этих книг распределение цен? (попробуйте нарисовать гистограмму)

```python
import matplotlib.pyplot as plt
import numpy as np

# Собираем данные для графиков
dist1 = df_classic[df_classic.price > 0].price
cnt_dist1 = len(df_classic[df_classic.price > 0])
mean_dist1 = round(df_classic[df_classic.price > 0].price.mean(),1)
median_dist1 = round(df_classic[df_classic.price > 0].price.median(),1)



plt.hist(dist1)
plt.title(f'''Классическая литература: кол-во={cnt_dist1}, средняя={mean_dist1}, медиана={median_dist1}''')

plt.xlabel("Цена")
plt.xlim([0, 10000])

plt.ylabel("Кол-во книг")
plt.ylim([0, 750])

plt.show()
```

Мне кажется лучше смотреть медиану по книгам, т.к. у некоторых книг кол-во страниц может быть очень высоким, что можно отнести к выбросам

### 3. Как соотносятся средние цены книг по Python, Машинному обучению и C++ ? (поясните, как именно вы доставали такие книги :), то есть "я буду доставать книги вот так, потому что...")
тут решено было построить 3 идентичных графика, друг под другом для более наглядного сравнения
Опять таки же из цен были убраны книги, по которым подгрузившаяся цена = 0

```python
import matplotlib.pyplot as plt
import numpy as np

# Делаем для каждой книги графики
dist1 = df_python[df_python.price > 0].price
cnt_dist1 = len(df_python[df_python.price > 0])
mean_dist1 = round(df_python[df_python.price > 0].price.mean(),1)
median_dist1 = round(df_python[df_python.price > 0].price.median(),1)

dist2 = df_ml[df_ml.price > 0].price
cnt_dist2 = len(df_ml[df_ml.price > 0])
mean_dist2 = round(df_ml[df_ml.price > 0].price.mean(),1)
median_dist2 = round(df_ml[df_ml.price > 0].price.median(),1)


dist3 = df_c_plus_plus[df_c_plus_plus.price > 0].price
cnt_dist3 = len(df_c_plus_plus[df_c_plus_plus.price > 0])
mean_dist3 = round(df_c_plus_plus[df_c_plus_plus.price > 0].price.mean(),1)
median_dist3 = round(df_c_plus_plus[df_c_plus_plus.price > 0].price.median(),1)



fig, axs = plt.subplots(3, 1, sharey=True, tight_layout=True,figsize=(10, 10))

# Параметры для графиков
axs[0].hist(dist1, color = '#1E90FF')
axs[0].set_title(f'''Книги по Python: кол-во={cnt_dist1}, средняя={mean_dist1}, медиана={median_dist1}''')
axs[0].set_ylabel("Кол-во книг")
axs[0].set_xlabel("Цена")
axs[0].set_xlim([0, 5000])
axs[0].set_ylim([0, 50])


axs[1].hist(dist2, color = '#006400')
axs[1].set_title(f'''Книги по ML: кол-во={cnt_dist2}, средняя={mean_dist2}, медиана={median_dist2}''')
axs[1].set_ylabel("Кол-во книг")
axs[1].set_xlabel("Цена")
axs[1].set_xlim([0, 5000])
axs[1].set_ylim([0, 50])

axs[2].hist(dist3, color = '#696969')
axs[2].set_title(f'''Книги по C++: кол-во={cnt_dist3}, средняя={mean_dist3}, медиана={median_dist3}''')
axs[2].set_ylabel("Кол-во книг")
axs[2].set_xlabel("Цена")
axs[2].set_xlim([0, 5000])
axs[2].set_ylim([0, 50])
```
